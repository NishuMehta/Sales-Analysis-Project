## LOAD DATA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('sales-data-cleaned.xlsx')
```

```python
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

#### EXPLORATORY DATA ANALYSIS

```python
total_sales = df["Sales"].sum()
print(f"Total Sales: ₹{total_sales}")
```

```python
# Sales Trend Over Time
df_grouped = df.groupby("Order Date")["Sales"].sum()
plt.figure(figsize=(10,5))
plt.plot(df_grouped, marker='o', color='g')
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Top 5 Best-Selling Products
top_products = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(8,5))
sns.barplot(x=top_products.index, y=top_products.values, palette="viridis", width=0.5)
plt.title("Top 5 Best-Selling Products")
plt.show()
```
